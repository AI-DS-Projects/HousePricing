## Missing values

```python
df.isnull().sum()
```

missing values present in N_BEDROOM,N_BATHROOM,QS_OVERALL

```python
df['N_BEDROOM'].isnull().sum()
```

```python
df['N_BEDROOM'].mode()
```

```python
df['N_BEDROOM'].fillna(value=(df['N_BEDROOM'].mode()[0]),inplace=True) #fill the missing value with the mode
```

```python
df['N_BEDROOM'].isnull().sum()
```

```python
df['N_BATHROOM'].isnull().sum() # 5 missing values in 'N_BATHROOM'
```

```python
df.loc[df['N_BATHROOM'].isnull()==True] #returns records where 'N_BATHROOM' is null
```

```python
#fill the bathroom with 1 if no of bedroom is 1 otherwise 2 
for i in range(0,len(df)):
    if pd.isnull(df['N_BATHROOM'][i])==True:
        if (df['N_BEDROOM'][i]==1.0):
            df['N_BATHROOM'][i]=1.0
        else:
            df['N_BATHROOM'][i]=2.0
```

```python
df['N_BATHROOM'].isnull().sum() # 5 missing values in 'N_BATHROOM'
```

```python
df['QS_OVERALL'].isnull().sum() #48 missing values in QS_OVERALL
```

```python
df[['QS_ROOMS','QS_BATHROOM', 'QS_BEDROOM', 'QS_OVERALL']].head()
```

```python
temp=(df['QS_ROOMS']+df['QS_BATHROOM']+df['QS_BEDROOM'])/3
pd.concat([df['QS_ROOMS'],df['QS_BATHROOM'],df['QS_BEDROOM'],df['QS_OVERALL'],temp],axis=1).head()
```

QS_OVERALL  approximatly  is average of QS_ROOMS,QS_BATHROOM,QS_BEDROOM

```python
def fillna(x):
    return ((x['QS_ROOMS'] + x['QS_BATHROOM'] + x['QS_BEDROOM'])/3)
```

```python
df['QS_OVERALL']=df.apply(lambda x: fillna(x) if pd.isnull(x['QS_OVERALL']) else x['QS_OVERALL'],axis=1)
```
